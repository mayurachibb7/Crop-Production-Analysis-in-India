## **Importing the Necessary Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Function for Data Preprocessing

```python
def data_preprocessing(path):
    # Load the dataset
    data = pd.read_csv(path)
    # Remove null values
    data_cleaned = data.dropna()
    # Display missing values
    print(data_cleaned.isnull().sum())
    # Display summary statistics
    print(data_cleaned.describe())
    return data_cleaned
```

### Function for plotting year wise crop production trend

```python
def Year_wise_crop_production_trend(data):
    plt.figure(figsize=(10, 5))
    sns.lineplot(x='Crop_Year', y='Production', data=data)
    plt.title('Year wise crop production trend')
    plt.xlabel('Year')
    plt.ylabel('Production')
    plt.show()
```

### Function for plotting crop wise crop production

```python
def crop_wise_crop_production(data):
    plt.figure(figsize=(20, 5))
    sns.barplot(x='Crop', y='Production', data=data, estimator=sum)
    plt.title('Crop wise crop production')
    plt.xlabel('Crop')
    plt.ylabel('Production')
    plt.xticks(rotation=90)
    plt.show()
```

### Function for plotting state wise crop production

```python
def state_wise_crop_production(data):
    plt.figure(figsize=(10, 5))
    sns.barplot(x='State_Name', y='Production', data=data, estimator=sum)
    plt.title('State wise crop production')
    plt.xlabel('State')
    plt.ylabel('Production')
    plt.xticks(rotation=90)
    plt.show()
```

### Function for plotting season wise crop production

```python
def season_wise_crop_production(data):
    plt.figure(figsize=(10, 5))
    sns.barplot(x='Season', y='Production', data=data, estimator=sum)
    plt.title('season_wise_crop_production')
    plt.xlabel('Season')
    plt.ylabel('Production')
    plt.show()
```

### Function for plotting distribution of crop production

```python
def distribution_of_crop_production(data):
    plt.figure(figsize=(10, 5))
    sns.histplot(data['Production'], bins=30, kde=True)
    plt.title('distribution_of_crop_production')
    plt.xlabel('Production')
    plt.ylabel('Count')
    plt.show()
```

```python
path = "/content/drive/MyDrive/Unified Mentor/Crop Production data.csv"
data_cleaned = data_preprocessing(path)
```
